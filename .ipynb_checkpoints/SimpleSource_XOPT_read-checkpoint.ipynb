```python
# import 
import sys 
import os 
import numpy as np 
import matplotlib.pyplot as plt 
from os.path import basename
from re import sub
import import_ipynb
from xopt.generators.ga.cnsga import CNSGAGenerator
from SimpleSource_XOPT_run import evaluate_TRAK_FUNCTION, TRAK_MK_vocs
from xopt import Xopt, Evaluator
```

```python
# define the evaluator
ev = Evaluator(function = evaluate_TRAK_FUNCTION)
ev.function_kwargs = {'raise_probability': 0.0}
```

```python
# define X
X = Xopt(
    generator = CNSGAGenerator(TRAK_MK_vocs),
    evaluator = ev,
    vocs = TRAK_MK_vocs
)
```

```python
# run range(#) different simulations in optimization
%%time 
for _ in range(20):
    X.step() 
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# plot the population of solutions

def plot_population(X):
   fig,ax = plt.subplots(figsize = (8, 8))
   
   fdata = TRAK_MK_vocs.feasibility_data(X.data)

   print(fdata)
   
   k1 = "an_x"
   k2 = "fe_x"
   
   X.data.plot.scatter(k1, k2, marker = ".", alpha = 0.2, color = "black", ax = ax)
   X.data[fdata["feasible"]].plot.scatter(k1, k2, marker = "x", alpha = 0.3, color = "orange", ax = ax)
   #X.generator.population.plot.scatter(k1, k2, marker = "o", color = "red", alpha = 1, ax = ax)
   ax.set_xlabel(k1)
   ax.set_ylabel(k2)
   ax.set_title("function optimization with Xopt's CNSGA")
```

```python
plot_population(X)
```
